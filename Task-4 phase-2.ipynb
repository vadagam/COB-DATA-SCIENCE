```python
set1=train_data[0]
set1.to_csv('train.csv')
train_data=pd.read_csv('train.csv')
train_data
```

```python
trains=train_data.drop("Unnamed: 0",axis=1)
trains.columns=trains.iloc[0]
train_data=trains.drop(0)
train_data
```

```python
set2=test_data[0]
set2.to_csv('test.csv')
test_data=pd.read_csv('test.csv')
test_data
```

```python
tests=test_data.drop("Unnamed: 0",axis=1)
tests.columns=tests.iloc[0]
test_data=tests.drop(0)
test_data
```

```python
train_dataset=train_data.dropna(subset=['y'])
train_dataset
```

```python
x_train=train_dataset[['x']].astype(float)
x_test=test_data[['x']].astype(float)
y_train=train_dataset[['y']].astype(float)
y_test=test_data[['x']].astype(float)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
yhat=model.predict(x_test)
yhat
```

```python
y_test=y_test.squeeze()
yhat=yhat.squeeze()
```

```python
showdata=pd.DataFrame({"Test_value":y_test, "Predicted_value":yhat})
showdata
```
