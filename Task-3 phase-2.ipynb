```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("dataset - netflix1.csv")
data.head()
```

```python
_df_=data.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
_df_.groupby('director').size().plot(kind='barh',color=sn.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top','right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
_df_['release_year'].plot(kind='hist', bins=20, title='release_year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data.info()
```

```python
plt.figure(figsize=(8,8))
sn.countplot(data=data,x='type',palette='Set1')
plt.xlabel('catogories',fontsize=10)
plt.ylabel('Counts',fontsize=10)
plt.title('movies vs tvshows',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,6))
sn.countplot(data=data,x='rating',order=data['rating'].value_counts().index,palette='Set2')
plt.xlabel('Rating',fontsize=10)
plt.ylabel('Counts',fontsize=10)
plt.title('Rating distribution',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,8))
top_countries=data['country'].value_counts().head(10)
sn.barplot(x=top_countries.values,y=top_countries.index,palette='muted')
plt.xlabel('count',fontsize=10)
plt.ylabel('Country',fontsize=10)
plt.title('countries with most shows',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,8))
sn.histplot(data=data,x='release_year',bins=30,kde=True,color='red')
plt.xlabel('year of release',fontsize=10)
plt.ylabel('Counts',fontsize=10)
plt.title('Release year distribution',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```

```python
plt.figure(figsize=(8,8))
genres=data['listed_in'].str.split(',').explode().str.strip()
top_genres=genres.value_counts().head(10)
sn.barplot(x=top_genres.values,y=top_genres.index,palette='dark:red')
plt.xlabel('count',fontsize=10)
plt.ylabel('Genre',fontsize=10)
plt.title('top genres',fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show
```
